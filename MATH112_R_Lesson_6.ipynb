8. **<font color= #C7B8EA;>How many of each "Type" and "Treatment" are there?</font>**

Because "conc" and "uptake" are quantitative variables, their summary is not a *count* but the

*    minimum,
*    first quartile,
*    median,
*    mean,
*    third quartile,
*    and maximum.

9. **<font color= #C7B8EA;>What insight into "conc" and "uptake" do we get from this summary?</font>**

```r
# let's save the mean of only "conc"
m = mean(CO2$conc)

# now let's determine which grass plants have less than the mean
# the next line returns the row numbers associated with desired condition
indx = which(CO2$conc < m)

# to get the rows, themselves
# but we only print the dimensions to see that the number of rows has changed
dim(CO2[indx, ])
```

```r
# I can also just highlight that I want ONLY chilled grass plants... for whatever reason
idx = which(CO2$Treatment == "chilled")
chilled = CO2[idx,]

# Let's separate the not chilled grass plants, because why not?
not_chilled = CO2[-idx, ]
```

Go ahead and explore the differences in "conc" and "uptake" when the two groups are separated.  

10. **<font color= #C7B8EA;>What do you find?</font>**

```r
# Now, we can put them back together, as we've learned before, using rbind or cbind.

# rbind needs the same number of columns.
# we see that the new dataset is similar to the original
rbind(chilled,not_chilled)

# cbind needs the same number of rows.
# we see that the function does NOT combine duplicated variables
cbind(chilled, not_chilled)
```
