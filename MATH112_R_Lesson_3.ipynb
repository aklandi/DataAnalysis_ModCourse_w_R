# II) **Casting Types**

You can switch object types as needed.  But this only makes sense to do for some variable relationships.  For example,

*   character <-> numeric
*   character <-> factor
*   combining basic object types into a data frame




Instance 1: when you have a vector of strings of numbers

```r
# create a vector
alpha = c("1", "2", "3", "4", "5")
# check its class
class(alpha)
```

```r
# cast as numeric
as.numeric(alpha)
```

Instance 2: when you have numbers you need to be strings

```r
# make vector
x = sample(1:100, 5, replace = FALSE)
# check class
class(x)
```

```r
# cast as strings
y = as.character(x)
class(y)
```

Instance 3: when you have a factor object but want only the strings, no levels

```r
# create a factor object
Fa = factor(c("male", "female", "nb"))
#nlevels(Fa)
class(Fa)
```

```r
# make factors strings
Ge = as.character(Fa)
Ge
```

```r
class(Ge)
```

Instance 4: when you have the strings, but need a factor with levels

```r
# make strings
gender = c("male", "female", "nb")
# cast as factors
as.factor(gender)
```
